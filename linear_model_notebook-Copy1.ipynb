```r
library(ggplot2)

# Data
df <- data.frame(
  x = c(1, 2, 3, 4, 5),
  y = c(2.1, 4.3, 6.1, 8.0, 10.1)
)

# Fit model
model <- lm(y ~ x, data = df)
slope <- coef(model)[2]
intercept <- coef(model)[1]
r <- cor(df$x, df$y)
pred <- predict(model)
mse <- mean((df$y - pred)^2)

# Plot
ggplot(df, aes(x = x, y = y)) +
  geom_point() +
  geom_smooth(method = "lm", se = FALSE, color = "red") +
  annotate("text", x = 1.5, y = max(df$y) - 0.5,
           label = paste("y =", round(slope, 2), "x +", round(intercept, 2),
                         "\nr =", round(r, 2), "\nMSE =", round(mse, 2)),
           size = 4) +
  labs(title = "Linear Fit",
       x = "Years of Experience", y = "Salary") +
  theme_minimal()

ggsave("regression_plot_r.png")
```

I added stats to display the slope, intercept, r-value, p-value, and standard error. The slope indicates that as you increase in Years of Experience, your salary increases. The intercept indicates that even with 0 years of experience, you still get some salary. The r-value indicates that the linear model is a good fit for the data. The MSE measures average squared error between predictions and actual values.
